```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the classes and subgroups
classes = ['Total P','P retention','Alkaline soil P','Soluble P']
total_p = ['TOT','TP',"stp"]
p_retention = ['occ','sol_VH','PNZ']
alkaline_p = ['POL','sol_Mo']
soluble_p = ['lab','PBR','PHO','extP']
subgroups = [total_p, p_retention, alkaline_p, soluble_p]
```

```python
# define the phosphorus classes and subgroups
classes = ['Total P', 'P Retention', 'Alkaline Soil P', 'Soluble P']
subgroups = [['tot', 'TP', 'stp'], ['occ', 'sol_VH', 'PNZ'], ['POL', 'sol_Mo'], ['lab', 'PHO', 'extP']]
```

```python
# Load the phenotype data for maize and sorghum
maize_pheno = pd.read_csv('phenotype/maize_phenotype.csv')
sorghum_pheno = pd.read_csv('phenotype/sorghum_phenotype.csv')
```

```python
print(maize_pheno.columns)
```

```python
# Define the figure and subplots
fig, axes = plt.subplots(nrows = len(classes), ncols=1, figsize=(8,12))
```

```python
# create a figure with subplots for each class
fig, axes = plt.subplots(nrows=len(classes), ncols=1, figsize=(8, 10), sharex=True)
```

```python
for i, c in enumerate(classes):
    axes[i].set_title(c)
    axes[i].set_ylabel('Phenotype value')
    for sg in subgroups[i]:
        maize_vals = maize_pheno[maize_pheno['Trait']==sg]['Value'].values
        sorghum_vals = sorghum_pheno[sorghum_pheno['Trait']==sg]['Value'].values
        axes[i].plot(maize_vals, label = 'Maize - {}'.format(sg))
        axes[i].plot(sorghum_vals, label = 'Sorghum - {}'.format(sg))
    axes[i].legend()
```

```python
# loop over each class
for i, c in enumerate(classes):
    axes[i].set_title(c)
    axes[i].set_ylabel('Phenotype value')
    # loop over each subgroup in the class
    for sg in ['lab', 'sol_VH', 'PHO']:
        maize_vals = maize_pheno[sg].values
        sorghum_vals = sorghum_pheno[sg].values
        plt.plot(maize_vals, label='Maize')
        plt.plot(sorghum_vals, label='Sorghum')
        plt.title(f'{sg} for Soluble P')
        plt.legend()
        plt.show()
```
